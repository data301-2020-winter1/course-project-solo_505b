### Importing the dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('~/Downloads/framingham.csv', sep =',')
df.head()
```

### 1.Preparing dataset
First of all I want to drop *education*column and next I'm going to rename the varibale *male* to *Sex* which 1 indicates being male and 0 indicates being female

```python
df.drop(['education'],axis=1,inplace=True) 
df.rename(columns={'male':'Sex'},inplace=True)
df.head()
```

Next step is to handle missing values

```python
df.dropna(axis=0,inplace=True)
```

Now, for handling outliers, we only need to consider the continous variables('totChol','sysBP','diaBP','BMI','hearRate','glucose'). one way to detect outliers is by looking at boxplot.

```python
plt.figure(figsize=(25,10))
sns.boxplot(data=df.iloc[:,[8,9,10,11,12,13]])
```

Boxplots are a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1/25th Percentile), median, third quartile (Q3/75th Percentile), and “maximum”).In order to detect outliers we can look at **InterQuartile Range (IQR)**: 25th to the 75th percentile. IQR tells how spread the middle values are. based on the IQR we can define min and max value as follows:<br> - “maximum”: Q3 + 1.5*IQR <br> - “minimum”: Q1 -1.5*IQR. <br> Ab outlier is an observation point that is distant from other observations. Based on the definition on min and max here, we consider any value beyond the min or max. Outliers detected in totChol and sysBP column, so we're going to remove them.

```python
maxtotChol = df['totChol'].max()
maxsysBP = df['sysBP'].max()
df=df[df['totChol']<maxtotChol]
df=df[df['sysBP']<maxsysBP]
df.to_csv('PreparedDataSet.csv')
```

Now with the missing values, outliers  dealt with, we proceed to perform EDA!

#### Research Question 1: what is the correlation between each colum?
A table displaying correlation coefficients between variables is called a correlation matrix. The correlation between two variables is presented by each cell in the table. A correlation matrix is used to summarize data. Correlation coefficients are indicators of the strength of the relationship between two different variables. A coefficient of correlation that is greater than zero suggests a positive relationship between two variables. A value that is less than zero means that two variables have a negative relationship. It implies that variation in one may explain a variation in another if two factors are strongly correlated. Often, strongly correlated variables may be excluded from the dataset. I'm going to use **heatmap** to display the correlations more efficiently.

#### Research Question 5: What are the most relevant features that affect target variable?
There are several methods to answer this question. Feature here simply means explanoritory variables or columns in the dataset. When we get a dataset, not necessarly every column whould have an effect on the target variable, as I mentioned in research question 1, if two features have correlation, presence of one of the them probably will suffice. I'm going to use **Backward Selection** method. The performance metric is  p-value, if the pvalue is bigger than 5 percent, feature should be removed, else we keep it. In order to apply backward selection method we do the evaluation in a loop over all features. At first, consider all features in the model. We review the performance of the model and then iteratively delete the worst performing features one by one until the average performance of the model is within a reasonable range. I used statsmodels package here

```python
import statsmodels.api as sm
X = df.iloc[:,0:14]
y = df.iloc[:,-1]
cols = list(X.columns)
pmax = 1
while (len(list(X.columns))>0):
    p= []
    X = X[cols]
    X = sm.add_constant(X)
    model = sm.OLS(y,X).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
relevant_features = cols
print(relevant_features)

```

- Top 5 high risk factors in this dataset are:
1. *systolic blood pressure*
2. *glucose*
3. *age*
4. *Sex*
5. *number of smoked cigarettes per day*
6. *history of heart attack*

#### Research Question 6: Performing logistic regression by using the most risk factors from previous research question!
Logistic Regression is used when the dependent variable(target) is categorical. Like all regression analyses, the logistic regression is a predictive analysis and is used to describe data and to explain the relationship between one dependent binary variable and one or more nominal, ordinal, interval independent variables. First, we need to split the dataset into training and testing parts. I used train_test_split form model_selection module in sklearn package. from the same package I used logistic regression to fit a model to my data. At the end, to compute accuracy of the fitted model, I used metrics.accuracy_score from the same package.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x=df[['sysBP','glucose','age','Sex','cigsPerDay','prevalentStroke']]
y=df.iloc[:,-1]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=5)
lreg=LogisticRegression()
lreg.fit(x_train,y_train)
y_hat=lreg.predict(x_test)
sklearn.metrics.accuracy_score(y_test,y_hat)
```
